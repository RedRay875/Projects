```python
import pandas as pd
from sklearn.utils import Bunch

# Load your dataset
house_price_dataset = pd.read_csv('C:/Users/KIIT0001/Desktop/ML/.ML/BostonHousing.csv')

# Separate features and target
X = house_price_dataset.drop(columns='price')  # Features
y = house_price_dataset['price']               # Target

# Wrap in a Bunch (like sklearn.datasets)
house_data_bunch = Bunch(
    data=X.values,
    target=y.values,
    feature_names=X.columns.tolist(),
    target_names=['price'],
    frame=house_price_dataset
)
```

## Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('BostonHousing.csv')

# Set style for plots
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))

# 1. Distribution of the target variable (price)
plt.figure(figsize=(10, 6))
sns.histplot(data['price'], kde=True, bins=30)
plt.title('Distribution of House Prices')
plt.xlabel('Price ($1000s)')
plt.ylabel('Frequency')
plt.show()

# 2. Correlation heatmap
plt.figure(figsize=(12, 10))
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Boston Housing Features')
plt.show()

# 3. Scatter plots of important features vs price
features_to_plot = ['rm', 'lstat', 'ptratio', 'crim', 'age']
for feature in features_to_plot:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=data[feature], y=data['price'])
    plt.title(f'{feature} vs Price')
    plt.xlabel(feature)
    plt.ylabel('Price ($1000s)')
    plt.show()

# 4. Boxplot of price by Charles River dummy variable (chas)
plt.figure(figsize=(8, 6))
sns.boxplot(x='chas', y='price', data=data)
plt.title('Price Distribution by Charles River Proximity')
plt.xlabel('Charles River (0 = No, 1 = Yes)')
plt.ylabel('Price ($1000s)')
plt.show()

# 5. Pairplot of selected features
selected_features = ['price', 'rm', 'lstat', 'ptratio', 'age']
sns.pairplot(data[selected_features])
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.show()

# 6. Price distribution by RAD (accessibility to radial highways)
plt.figure(figsize=(12, 6))
sns.boxplot(x='rad', y='price', data=data)
plt.title('Price Distribution by Accessibility to Radial Highways')
plt.xlabel('Index of Accessibility to Radial Highways')
plt.ylabel('Price ($1000s)')
plt.show()

# 7. Crime rate vs price with different colors for high and low crime
plt.figure(figsize=(10, 6))
sns.scatterplot(x='crim', y='price', hue=data['crim']>1, data=data, palette=['blue','red'])
plt.title('Crime Rate vs Price')
plt.xlabel('Crime Rate')
plt.ylabel('Price ($1000s)')
plt.legend(title='High Crime', labels=['No', 'Yes'])
plt.show()
```

## Model training 

### XGBoost Regressor

```python
model = XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

## Evalution

### Prediction of Data

```python
train_prediction = model.predict(X_train)

print("Training data :",train_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train,train_prediction)

# Main squared error
score_2 = metrics.mean_absolute_error(Y_train,train_prediction)

print("R scored error :",score_1)
print("Main scored error :",score_2)
```

```python
plt.scatter(Y_train,train_prediction)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Price')
plt.show()
```

```python
test_prediction = model.predict(X_test)
```

```python
score_3 = metrics.r2_score(Y_test,test_prediction)

# Main squared error
score_4 = metrics.mean_absolute_error(Y_test,test_prediction)

print("R scored error :",score_3)
print("Main scored error :",score_4)
```
